```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('font',size=14)
e=0.0010
K=3.16668935659E-6
T=573.15
B=1/(K*T)
q=1/(1-np.exp(-B*e))
ei=np.arange(0.0,0.01,0.00001)
pi=np.exp(-B*ei)/q
fig,ax=plt.subplots()
ax.plot(pi,ei,'b',linewidth=3)
ax.set(xlabel='Population',ylabel='Energy (Hartree)',
    title='T = %f, q = %f' % (T,q))
for i in range(int(0.01/e)+1):
    ax.plot([0,1],[i*e,i*e],'k')
    rect = matplotlib.patches.Rectangle((0,i*e),
        np.exp(-B*i*e)/q,0.8*e,linewidth=2,edgecolor='k',
        facecolor='r')
    ax.add_patch(rect)
```
